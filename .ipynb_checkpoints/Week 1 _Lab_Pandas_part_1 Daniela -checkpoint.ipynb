### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
## Some car brand names refer to the same brand. Replace all brand names that contain the word 
##"Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use loc with boolean indexing.

vehicles_df[vehicles_df["Make"].str.contains("Dutton")]
```

```python
# vehicles_df["Make"].value_counts()
```

```python
#vehicles_df = vehicles_df[vehicles_df["Make"].str.contains("Dutton")] = "Dutton"
```

```python
vehicles_df.iloc[vehicles_df["Make"].str.contains("Dutton"),0] = "Dutton"
vehicles_df[vehicles_df["Make"].str.contains("Dutton")]
```

```python
vehicles_df
```

```python
#Convert CO2 Emissions from Grams/Mile to Grams/Km

vehicles_df['CO2 Emission Grams/Mile'] = vehicles_df['CO2 Emission Grams/Mile'] / 1.60934 
vehicles_df.rename(columns={"CO2 Emission Grams/Mile" : "CO2 Emission Grams/Km"}, inplace=True)


vehicles_df
```

```python
#Create a binary column that solely indicates if the transmission of a car is automatic or manual. 
#Use pandas.Series.str.startswith and .

bool_automatic = vehicles_df["Transmission"].str.startswith("Auto", na = False)
bool_automatic

vehicles_df ['Auto/Manual'] = bool_automatic
vehicles_df 
```

```python
#data_new1 = data.copy()                                                # Create copy of pandas DataFrame
#data_new1['x1'] = data_new1['x1'].map({True: 'True', False: 'False'})  # Replace boolean by string
#print(data_new1) 
# https://statisticsglobe.com/convert-boolean-string-pandas-dataframe-column-python


vehicles_df["Auto/Manual"] = vehicles_df["Auto/Manual"].map({True:"Automatic", False:"Manual"})
vehicles_df
```

```python
vehicles_df['City MPG'] = vehicles_df['City MPG'] / 2.352
vehicles_df['Highway MPG'] = vehicles_df['Highway MPG'] / 2.352
vehicles_df['Combined MPG'] = vehicles_df['Combined MPG'] / 2.352
vehicles_df.rename(columns={'City MPG' : 'City km_per_liter',"Highway MPG" : 'Highway km_per_liter', "Combined MPG" : 'Combined km_per_liter'}, inplace=True)

vehicles_df
```
