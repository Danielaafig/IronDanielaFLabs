```python
import random

def sns_hist_plot(numerical):
    colors=["coral","red","chocolate","bisque","gold","yellowgreen",
            "lawngreen","lightgreen","darkgreen","cyan","dodgerblue",
            "royalblue","mediumpurple","darkviolet","magenta","deeppink","pink"]
    col_num=len(list(numerical.columns))
    
    if col_num<4:
        fig1, axes = plt.subplots(1,col_num,figsize=(5*col_num,6))
    else:
        fig1, axes = plt.subplots(int(np.ceil(col_num/4)),4,figsize=(16,1.2*col_num))
    sns.set(style="darkgrid")
    
    for i,column in enumerate(numerical.columns):
        if col_num<=4:
            sns.histplot(x=column, kde=True, data=numerical, ax=axes[0,i],color=random.choice(colors))
        else:
            fig1=sns.histplot(x=column, kde=True, data=numerical, ax=axes[int((i//4)),i%4],color=random.choice(colors))
    
    return fig1
```

```python
plot_geral = sns_hist_plot(numerical)
plot_geral
```

```python
import random

def sns_hist_plot_2(numerical):
    colors=["coral","red","chocolate","bisque","gold","yellowgreen",
            "lawngreen","lightgreen","darkgreen","cyan","dodgerblue",
            "royalblue","mediumpurple","darkviolet","magenta","deeppink","pink"]
    col_num=len(list(numerical.columns))
    # Define the number of rows and columns you want
    n_rows=2
    n_cols=4
    fig3, axes = plt.subplots(nrows = n_rows, ncols = n_cols, figsize=(20,10))
    
    for i, column in enumerate(numerical.columns):
        fig3 = sns.histplot(numerical[column],ax=axes[i//n_cols,i%n_cols],color=random.choice(colors))
    
    return fig3
```
